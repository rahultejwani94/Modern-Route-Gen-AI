###### https://scikit-learn.org/0.15/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents = ["people watch movie and watch movie again", 
             "people watch cricket and watch cricket",
             "people like movie and like movie a lot",
             "people like cricket"]
```

```python
document=["divesh is a good mentor he teaches us very well he teaches the project"]
```

AIM: Covert: Data: Number

```python
bow = CountVectorizer()
```

```python
bow_matrix = bow.fit_transform(documents)
```

```python
bow_matrix
```

```python
print("BOW Matrix:\n", bow_matrix.toarray())
```

```python
bow.get_feature_names_out()
```

```python
print("Vocabulary:", bow.get_feature_names_out())
```
