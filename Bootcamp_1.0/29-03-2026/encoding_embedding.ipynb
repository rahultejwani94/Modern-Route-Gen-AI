Encoding: representation of data in numbers(using vector)

Embedding: representation of meaningful data in numbers(using vector)

# Encoding
- encoding doesn't prevent semantic meaning of the data
- Encoding is count based technique
- Encoding -> data, text, image, video -> vector [1,2,3,...]
- Types of Encoding Techniques:
    - one hot encoding
    - BOW (bag of words)
    - TF_IDF
    - BM25 (ranking)

# Embedding
- Embedding and vectors are same (in maths we call vector, in NLP, GenAI we call them embedding)  
- It contains meaningful data  
- Embedding -> data, text, image, video -> vector [1,2,3,...]
- Embedding is not count based
- whatever embedding techniques are used nowadays are SOTA(state of the art model) based on Transformer. These models are capable of converting words, sentences, paragraphs, documents, book, image, video any kind of data into embedding.
- Early embedding methods used in 2018, 2019, 2020 were based on Neural Network. These models were capable of converding only word/text data into embedding
    - Word2Vec
    - FastText
    
## Terms
- Character -> smallest unit (A,B,C...)
- Token -> collection of characters
- word -> may split into tokens
- sentence -> may split into words/tokens
- paragraph -> may split into sentences
- document -> may split into paragraphs
- book -> may split into documents
